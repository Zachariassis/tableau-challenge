```python
import pandas as pd
```

```python
bike19Jan=pd.read_csv('Resources/201901-citibike-tripdata.csv').dropna()
bike19Feb=pd.read_csv('Resources/201902-citibike-tripdata.csv').dropna()
bike19Mar=pd.read_csv('Resources/201903-citibike-tripdata.csv').dropna()
bike19Apr=pd.read_csv('Resources/201904-citibike-tripdata.csv').dropna()
bike19May=pd.read_csv('Resources/201905-citibike-tripdata.csv').dropna()
bike19Jun=pd.read_csv('Resources/201906-citibike-tripdata.csv').dropna()
bike19Jul=pd.read_csv('Resources/201907-citibike-tripdata.csv').dropna()
bike19Aug=pd.read_csv('Resources/201908-citibike-tripdata.csv').dropna()
bike19Sep=pd.read_csv('Resources/201909-citibike-tripdata.csv').dropna()
bike19Oct=pd.read_csv('Resources/201910-citibike-tripdata.csv').dropna()
bike19Nov=pd.read_csv('Resources/201911-citibike-tripdata.csv').dropna()
bike19Dec=pd.read_csv('Resources/201912-citibike-tripdata.csv').dropna()
```

```python
bike19Jan['Month']='Jan'
bike19Feb['Month']='Feb'
bike19Mar['Month']='Mar'
bike19Apr['Month']='Apr'
bike19May['Month']='May'
bike19Jun['Month']='Jun'
bike19Jul['Month']='Jul'
bike19Aug['Month']='Aug'
bike19Sep['Month']='Sep'
bike19Oct['Month']='Oct'
bike19Nov['Month']='Nov'
bike19Dec['Month']='Dec'
```

```python
biketotal=pd.concat([bike19Jan, bike19Feb, bike19Mar, bike19Apr, bike19May, bike19Jun, bike19Jul, bike19Aug, bike19Sep, bike19Oct, bike19Nov, bike19Dec], ignore_index=True)
del bike19Jan
del bike19Feb
del bike19Mar
del bike19Apr
del bike19May
del bike19Jun
del bike19Jul
del bike19Aug
del bike19Sep
del bike19Oct
del bike19Nov
del bike19Dec
```

```python
biketotal=biketotal.loc[biketotal['gender']>0]
biketotal=biketotal.loc[biketotal['birth year']>1968]
biketotal=biketotal.loc[biketotal['usertype']=='Subscriber']
```

```python
biketotal.head()
```

```python
d=[biketotal['start station name'], biketotal['end station name'], biketotal['Month']]
bikelocations = pd.concat(d, axis=1, keys=['start station name', 'end station name', 'Month'])
bikelocations.head()
```

```python
d = [biketotal['end station name'], biketotal['end station latitude'], biketotal['end station longitude']]
bikeloctations_full = pd.concat(d, axis=1, keys=['end station name', 'end station latitude','end station longitude'])

latlong = bikeloctations_full.groupby(['end station name']).mean().rename(columns={'end station latitude':'latitude', 'end station longitude': 'longitude'})
latlong
```

```python
Jan=bikelocations.loc[bikelocations['Month']=='Jan']
start = Jan.groupby(['start station name']).count()
bikecounts_start = start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Jan'})
del Jan

Feb=bikelocations.loc[bikelocations['Month']=='Feb']
start = Feb.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Feb'}), how='outer')
del Feb

Mar=bikelocations.loc[bikelocations['Month']=='Mar']
start = Mar.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Mar'}), how='outer')
del Mar

Apr=bikelocations.loc[bikelocations['Month']=='Apr']
start = Apr.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Apr'}), how='outer')
del Apr

May=bikelocations.loc[bikelocations['Month']=='May']
start = May.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start May'}), how='outer')
del May

Jun=bikelocations.loc[bikelocations['Month']=='Jun']
start = Jun.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Jun'}), how='outer')
del Jun

Jul=bikelocations.loc[bikelocations['Month']=='Jul']
start = Jul.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Jul'}), how='outer')
del Jul

Aug=bikelocations.loc[bikelocations['Month']=='Aug']
start = Aug.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Aug'}), how='outer')
del Aug

Sep=bikelocations.loc[bikelocations['Month']=='Sep']
start = Sep.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Sep'}), how='outer')
del Sep

Oct=bikelocations.loc[bikelocations['Month']=='Oct']
start = Oct.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Oct'}), how='outer')
del Oct

Nov=bikelocations.loc[bikelocations['Month']=='Nov']
start = Nov.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Nov'}), how='outer')
del Nov

Dec=bikelocations.loc[bikelocations['Month']=='Dec']
start = Dec.groupby(['start station name']).count()
bikecounts_start = bikecounts_start.join(start.drop(['Month'], axis = 1).rename(columns={'end station name':'counts start Dec'}), how='outer')
del Dec
del start

bikecounts_start=bikecounts_start.fillna(0).astype(int)
bikecounts_start['counts start Sum'] = bikecounts_start['counts start Jan'] + \
            bikecounts_start['counts start Feb'] + bikecounts_start['counts start Mar'] + \
            bikecounts_start['counts start Apr'] + bikecounts_start['counts start May'] + \
            bikecounts_start['counts start Jun'] + bikecounts_start['counts start Jul'] + \
            bikecounts_start['counts start Aug'] + bikecounts_start['counts start Sep'] + \
            bikecounts_start['counts start Oct'] + bikecounts_start['counts start Nov'] + \
            bikecounts_start['counts start Dec']

bikecounts_start
```

```python
bikecounts = bikecounts_start.join(bikecounts_end, how='outer').fillna(0).astype(int)
bikecounts = bikecounts.join(latlong, how='outer')
bikecounts
```
